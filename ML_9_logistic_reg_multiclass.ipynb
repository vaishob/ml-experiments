## The Digit Dataset

- In this tutorial, we will be using the `load_digits` library/dataset from `sklearn`. This dataset is made up of 1797 8x8 images. Each image, like the one shown below, is of a hand-written digit. In order to utilize an 8x8 figure like this, we'd have to first transform it into a feature vector with length 64.

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import numpy as np
import seaborn as sn
```

```python
# Load the digits dataset containing images of handwritten digits
digits = load_digits()
```

```python
# List all attributes and methods of the digits dataset object
dir(digits)
```

```python
# Attributes of the digits dataset:
# DESCR - Description of the dataset
# data - Flattened pixel data for each image
# feature_names - Names of the features (if available)
# frame - DataFrame format of the data (if available)
# images - Original 8x8 pixel images
# target - Actual digit labels
# target_names - Unique digit classes (0-9)
```

```python
# Retrieve and display the pixel values for the first digit image in the dataset
digits.data[0]
```

```python
# Set the colormap to grayscale and display the first 5 digit images from the dataset
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
# Retrieve the labels for the first 5 digit images (shown above) in the dataset
digits.target[0:5]
```

```python
# Split the digits dataset into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# Create an instance of the LogisticRegression model for classifying the digits dataset
model = LogisticRegression()
```

```python
# Train the LogisticRegression model using the training data
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Display the 68th digit image from the dataset as an 8x8 grid
plt.matshow(digits.images[67])
```

```python
# Retrieve the actual digit label for the 68th image in the dataset
digits.target[67]
```

```python
# Predict the digit represented by the 68th image in the dataset using the trained model
model.predict([digits.data[67]])
```

```python
# Predict the digits represented by the first 5 images in the dataset using the trained model
model.predict(digits.data[0:5])
```

```python
# Generate predictions for the test set using the trained logistic regression model
y_predicted = model.predict(X_test)
```

```python
# Generate the confusion matrix to evaluate the performance of the model on the test set
cm = confusion_matrix(y_test, y_predicted)
cm
```

### Confusion Matrix Interpretation

The confusion matrix below shows the classification performance for digits 0-9:

- **Correct Predictions**: The diagonal elements represent the correctly classified images.
  - Examples:
    - 37 images of digit '0' were correctly classified as '0'.
    - 33 images of digit '1' were correctly classified as '1'.
    - 39 images of digit '2' were correctly classified as '2'.

- **Misclassifications**: The off-diagonal elements indicate where the model made errors.
  - Examples:
    - 1 image of digit '1' was incorrectly classified as '8'.
    - 1 image of digit '4' was incorrectly classified as '9'.
    - 2 images of digit '8' were misclassified, with one as '1' and another as '5'.

### Summary

- The model generally performed well, with most predictions correctly aligning along the diagonal.
- Some digits, like '8', showed more frequent misclassifications, suggesting possible areas for model improvement.

```python
# Plot a heatmap of the confusion matrix using Seaborn, and label the axes
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

## Exercise

Use `sklearn.datasets` iris flower dataset to train your model using logistic regression. You need to figure out accuracy of your model and use that to predict different samples in your test dataset. in iris dataset there are 150 samples containing following features:

* `Sepal Length`
* `Sepal Width`
* `Petal Length`
* `Petal Width`

Using above 4 features, you will classify a flower in one of the three categories:

* **Setosa**
* **Versicolour**

```python
# Load the iris dataset
iris = load_iris()
```

```python
X = iris.data
```

```python
y = iris.target
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize the logistic regression model
model = LogisticRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
y_pred
```

* `0` represents **Setosa** 
* `1` represents **Versicolour**
* `2` represents **Virginica**

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

The logistic regression model achieved an accuracy of **100%** on the test set.
